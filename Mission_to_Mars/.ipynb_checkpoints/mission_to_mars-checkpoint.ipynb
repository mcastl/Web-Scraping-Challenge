```python
# Import dependecies
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
#Set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
# Visit nasa website
url = 'https://redplanetscience.com/'
browser.visit(url)
html = browser.html
time.sleep(1)
#Scrape page into soup
soup = bs(html, "html.parser")
print(soup.prettify())
```

```python
#retrieving the latest news title and paragraph text
news_title = soup.find(class_='content_title').text
p_text = soup.find('div', class_= 'article_teaser_body')
print(news_title)
print(p_text)
```

```python
#visiting the URL for the JPL Featured Space Image through splinter
jpli_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(jpli_url)
# HTML object
html=browser.html
# Parse HTML
soup=bs(html,"html.parser")
sub_url= soup.find('div',class_='carousel_items')('article')[0]['style'].\
replace('background-image: url(','').replace(');','')[1:-1]
jpl_url = 'https://www.jpl.nasa.gov'
#Bild the full url
featured_url=jpl_url + sub_url
print(featured_url)
```
