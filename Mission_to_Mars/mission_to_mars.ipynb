```python
# Import dependecies
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
#Set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
# Visit nasa website
url = 'https://redplanetscience.com/'
browser.visit(url)
html = browser.html
time.sleep(1)
#Scrape page into soup
soup = bs(html, "html.parser")
```

```python
#retrieving the latest news title and paragraph text
news_title = soup.find(class_='content_title').text
p_text = soup.find('div', class_= 'article_teaser_body').text
print(news_title)
print(p_text)
```

```python
#Visit the url for the Featured Space Image site 
image_url = 'https://spaceimages-mars.com'
browser.visit(image_url)
print(soup.prettify())
```

```python
#Use splinter to navigate the site and find the image url
# HTML object
image_html=browser.html
# Parse HTML
image_soup=bs(image_html,"html.parser")

#retreieve the sub-url to the background image
sub_url= image_soup.find("carousel-item")["style"].replace('background-image: url(','').replace(');', '')[1:-1]

#creating the full url 
featured_image_url = image_url + sub_url
    
#print results
print(featured_image_url)
```
